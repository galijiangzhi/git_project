```python
import pandas as pd
import numpy as np
```

```python
list = pd.read_excel('./作者编号列表2.0.xlsx')
```

```python
list
```

```python
organization_list=list.loc[:,'organization']
```

```python
organization=pd.DataFrame(data=organization_list.drop_duplicates())
```

```python
organization.insert(loc=0,column='number',value='')
```

```python
organization.reset_index(drop=True,inplace=True)
```

```python
for i in range(len(organization)):
    organization.loc[i,'number']=i+1
```

```python
organization
```

```python
organization.to_excel('./excel/大学编号列表.xlsx')
```
