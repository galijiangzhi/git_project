```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('./excel/data.xlsx')
authors = pd.read_excel('./excel/作者关系全中文关系/’个人-高校‘中文片段关系表_数字版.xlsx')
journal = pd.read_excel('./excel/journal2.0.xlsx')
```

```python
authors.head(2)
```

```python
journal.head(2)
```

```python
df = pd.DataFrame(data=0,index=authors.loc[:,'authors'],columns=journal.loc[:,'journal'])
```

```python
df
```

```python
for i in range(len(data)):
    item = data.loc[i+1]
    try:
        author_id_list = str(item.loc['authors_id']).split('-')
        if '' in author_id_list:
            author_id_list.remove('')
        for j in range(len(author_id_list)):
            a = authors.loc[int(author_id_list[j])-1,'authors']
            df.loc[a,item.loc['journal_plus']]+=1
    except Exception as e:
        pass
```

```python
df
```

```python
df.to_excel('./excel/作者-报刊二维关系表.xlsx')
```
