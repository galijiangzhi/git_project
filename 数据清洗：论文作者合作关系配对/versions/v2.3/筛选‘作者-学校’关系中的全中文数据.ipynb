```python
authors_list=pd.read_excel('./excel/作者编号列表2.0.xlsx')
```

```python
authors_cp_list=pd.read_excel('./excel/作者关系全中文关系-姓名版.xlsx')
```

```python
authors_cp_list['author_a'][0]
```

```python
authors_list.insert(loc=0,column='result',value='False')
```

```python
for i in range(len(authors_cp_list)):
    str1=authors_cp_list.loc[i,'author_a']
    str2=authors_cp_list.loc[i,'author_b']
    for j in range(len(authors_list)):
        if authors_list.loc[j,'authors']==str1:
            authors_list.loc[j,'result']=True
            break
    for k in range(len(authors_list)):
        if authors_list.loc[k,'authors']==str2:
            authors_list.loc[k,'result']=True
```

```python
authors_plus=authors_list.loc[authors_list['result']==True]
```

```python
authors_plus.reset_index(inplace=True)
```

```python
authors_plus.drop('index',axis=1,inplace=True)
```
