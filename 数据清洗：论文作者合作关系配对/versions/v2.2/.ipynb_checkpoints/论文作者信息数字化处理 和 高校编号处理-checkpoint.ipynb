```python
def fuhao(data):
    return data.replace(' ','').replace(',','').replace('，','').replace(';','').replace('；','').replace('.','').replace('·','').replace('-','')
```

```python
def cn(data):#判断中文
    res = True
    for w in data.replace(' ','').replace(',','').replace('，','').replace(';','').replace('；','').replace('.','').replace('·',''):
        if not '\u4e00' <= w <= '\u9fff':
            res = False
            break
    return res
```

```python
def name(data):
    result1 = py.get_pinyin(data).upper()
    s = result1.split('-')
    s1 = s[0]
    del s[0]
    result2=s1+''.join(i[0] for i in s)
    return result2
```

```python
def cppd(str1,str2):
    if cn(str1):#如果这次操作的是一个中文
        if cn(str2):#str1和str2都是中文
            if (str1.replace(' ','').replace('.',''))==(str2.replace(' ','').replace('.','')):
                return True
            else:
                return False
        else:
            str1_py = py.get_pinyin(str1).upper().replace('-','')#全名拼音
            str1_py_s = name(str1)#拼音名缩写
            if (string_similar(fuhao(str1_py), fuhao(str2))>0.89) or (string_similar(fuhao(str1_py_s), fuhao(str2))>=0.9):
                return True
            else:
                return False
    else:#str1是一个英文
        str1_py = py.get_pinyin(str1).upper().replace('-','')#全名拼音
        str1_py_s = name(str1)#拼音名缩写
        str2_py = py.get_pinyin(str2).upper().replace('-','')#全名拼音
        str2_py_s = name(str2)
        if cn(str2):#str2是中文:
            #print(f'{str1}和{str2}使用en,en')
            if (string_similar(fuhao(str2_py), fuhao(str1_py))>0.89) or (string_similar(fuhao(str2_py_s), fuhao(str1_py))>=0.9):
                #print(f'{str1}和{str2}的相似度是{string_similar(fuhao(str2_py), fuhao(str1_py))}和{string_similar(fuhao(str2_py_s), fuhao(str1_py))}')
                return True
            else:
                return False
        else:#两个都是英文
            #print(f'{str1}和{str2}使用en,cn')
            if (string_similar(fuhao(str2_py),fuhao(str1_py))>0.89):
                return True
            else:
                return False
```

```python
def chinese0_pro(data):
    lista=data.split(' ')
    if '等' in lista:
        del lista[-1]
    for i in range(len(lista)):
        if i < len(lista):
            if len(lista[i])==1:
                if len(lista[i+1])==1:
                    lista[i]+=lista[i+1]
                    del lista[i+1]
    return lista
def chinese1_pro(data):
    lista=data.replace(' ','').replace('；',';').replace(',',';').replace('，',';').split(';')
    return lista

def chinese2_pro(data):
    lista=data.replace('；',';').replace(',',';').replace('，',';').split(';')
    return lista

def english0and1_pro(data):
    lista=data.split(';')
    return lista

def english0_1_pro(data):
    lista=data.split(';')
    listb=[]
    a=False
    ak = 0
    while '' in lista:
        lista.remove('')
    while ak<len(lista):
        a=(len(lista[ak])<=5)
        if a==True:#存在小于五个的情况
            for i in range((int)(len(lista)/2)):
                listb.append(lista[2*i-2]+lista[2*i-1]) 
                lista=listb
                break
        ak+=1   
    return lista

def english2_pro(data):#只包含空格和逗号的英文
    lista=data.split('，')
    listb=[]
    a=False
    ak = 0
    while '' in lista:
        lista.remove('')
    while ak<len(lista):
        a=(len(lista[ak])<=8)
        if a==True:#存在小于五个的情况
            for i in range((int)(len(lista)/2)):
                listb.append(lista[2*i-2]+lista[2*i-1]) 
                lista=listb
                break
        ak+=1
    return lista

def english3_pro(data):#只包括空格-;
    lista=data.split(';')
    listb=[]
    a=False
    ak = 0
    while '' in lista:
        lista.remove('')
    while ak<len(lista):
        a=(len(lista[ak])<=5)
        if a==True:#存在小于五个的情况
            for i in range((int)(len(lista)/2)):
                listb.append(lista[2*i-2]+lista[2*i-1]) 
                lista=listb
                break
        ak+=1
    return lista

def english4_pro(data):
    lista=data.split(';')
    return lista


def english5_pro(data):
    lista=data.replace('.;',';').replace('.,',';').replace('.，',';').replace('and',';').split(';')
    if 'Pand' in data:
        lista=data.replace('.;',';').replace('.,',';').replace('.，',';').split(';')
    if data == df.loc[2782]["authors"] or data == df.loc[6480]["authors"]:
        lista=data.replace(',;',',').split(',')
    elif data == df.loc[2878]["authors"]:
        lista=data.split('.;')
    elif data == df.loc[2879]["authors"]:
        lista=['Zhou;J. L.','Wang X','Zhang J.F.','Wang H.']
    while '' in lista:
        lista.remove('')
    return lista

def english6_pro(data):
    lista=data.replace('and',';').split(';')
    if 'Pand' in data or 'uand' in data:
        lista=data.split(';')
    if ';' not in data:
        if len(data)<11:
            lista=data.split(';')
        else:
            lista=data.split(',')
    if data == df.loc[7132]["authors"]:
        lista=data.replace(',',';').split(';')
    if data == df.loc[6241]["authors"]:
        lista=data.split(';')
    return lista

def english7_pro(data):
    lista=data.split(';')
    return lista

def english8_pro(data):
    lista=data.split(';')
    listb=[]
    a=False
    ak = 0
    while '' in lista:
        lista.remove('')
    while ak<len(lista):
        a=(len(lista[ak])<5)
        if '.' in lista[ak]:
            a == False 
        if a==True:#存在小于五个的情况
            
            for i in range((int)(len(lista)/2)):
                listb.append(lista[2*i-2]+lista[2*i-1]) 
                lista=listb
                break
        ak+=1
    return lista

def english9_pro(data):
    lista=data.replace('and',',').split(',')
    return lista

def english10_pro(data):
    lista=data.split('，')
    return lista

def english11_pro(data):
    lista=data.split('，')
    return lista

def english12_pro(data):
    lista=data.split(',')
    return lista

def english13_pro(data):
    lista=data.split(';')
    return lista

def english14_pro(data):
    lista=data.split('，')
    return lista

def english15_pro(data):
    lista=data.split('，')
    return lista

def lang0_0_pro(data):
    lista=data.replace('；',';').split(';')
    if data == '赵春晖 Wei Wang':
        lista=['赵春晖', 'Wei Wang']
    if data == '祝乔;Yin;Yi-Xin;Cheng;Han-Qing;Chen;Xian-Zhong':
        lista=['祝乔','Yin Yi-Xin','Cheng Han-Qing','Chen;Xian-Zhong']
    return lista

def lang0_1_pro(data):
    lista=data.split('，')
    return lista

def lang0_2_pro(data):
    lista=data.replace(',',';').split(';')
    return lista

def lang0_pro(data):
    lista=data.replace('第一通讯作者','').split(';')
    return lista
def lang1_pro(data):
    lista=data.replace('li&','').split(';')
    if data == 'Cheng;Y、Mao;JF、Wu;AH 、 Wu;GQ;Mao Jingfeng;Wu Aihua;Wu Guoqing':
        lista=['ChengY','MaoJF','WuAH','WuGQ','Mao Jingfeng','Wu Aihua','Wu Guoqing']
    return lista

def lang2_pro(data):
    if data == df.loc[4175,"authors"]:
        lista=['Yajun Zhang','Y. Jia','T. Chai','D. Wang','W. Dai','J. Fu']
    if data == df.loc[4176,"authors"]:
        lista=['Yajun Zhang','T. Chai','H. Wang','D. Wang','X. Chen']
    if data == df.loc[4556,"authors"]:
        lista=['Xiaokun Dai','Yang Song','Mira Schüller','Dieter Schramm']
    if data == df.loc[4612,"authors"]:
        lista=['Xianzhen Zhang','Zuohua Liu','Hui Peng','Xianmin Zhan']
    if data == df.loc[4617,"authors"]:
        lista=['Xianmin Zhang','Praveen Agarwal','Zuohua Liu','Hui Peng']
    if (data == df.loc[5924,"authors"]) or (data == df.loc[5925,"authors"]):
        lista=['Qi Lü']
    if data == df.loc[6808,"authors"]:
        lista=['Kwon,Joseph Sang-Il','Nayhouse,Michael','Orkoulas,Gerassimos','Ni, Dong','Christofides, Panagiotis D.']
    if data == df.loc[8366,"authors"]:
        lista=['Crose, Marquis','Kwon, Joseph Sang-Il','Nayhouse, Michael','Ni, Dong','Christofides, Panagiotis D.']    
    return lista

def authors_pro(data,authorstype):
    if authorstype=="chinese0":#只包括空格的中文
        return chinese0_pro(data)
    elif authorstype=="chinese1":#只包含符号;；,·，的中文
        return chinese1_pro(data)
    elif authorstype=="chinese2":#只包含符号;；,·，.的中文
        return chinese2_pro(data)
    elif authorstype=="english0":#只包括空格的英文
        return english0and1_pro(data)
    elif authorstype=="english0_1":#只包括空格和分号的英文
        return english0_1_pro(data)
    elif authorstype=="english1":
        return english0and1_pro(data)
    elif authorstype=="english2":
        return english2_pro(data)
    elif authorstype=="english3":
        return english3_pro(data)
    elif authorstype=="english4":
        return english4_pro(data)
    elif authorstype=="english5":
        return english5_pro(data)
    elif authorstype=="english6":
        return english6_pro(data)
    elif authorstype=="english7":
        return english7_pro(data)
    elif authorstype=="english8":
        return english8_pro(data)
    elif authorstype=="english9":
        return english9_pro(data)
    elif authorstype=="english10":
        return english10_pro(data)
    elif authorstype=="english11":
        return english11_pro(data)
    elif authorstype=="english12":
        return english12_pro(data)
    elif authorstype=="english13":
        return english13_pro(data)
    elif authorstype=="english14":
        return english14_pro(data)
    elif authorstype=="english15":
        return english15_pro(data)
    elif authorstype=="lang0_0":
        return lang0_0_pro(data)
    elif authorstype=="lang0_1":
        return lang0_1_pro(data)
    elif authorstype=="lang0_2":
        return lang0_2_pro(data)
    elif authorstype=="lang0":
        return lang0_pro(data)
    elif authorstype=="lang1":
        return lang1_pro(data)
    elif authorstype=="lang2":
        return lang2_pro(data)
```
